## <center>**Osilasi harmonik**</center> 
Osilasi harmonik adalah gerakan periodik di mana suatu sistem berosilasi (bergetar) bolak-balik di sekitar posisi kesetimbangannya. Sistem osilasi harmonik ideal ditandai oleh gaya pemulih yang sebanding dengan perpindahan dari posisi kesetimbangan, sesuai dengan hukum Hooke:
$$ F = -kx $$
Dimana:
- F : gaya pemulihan
- k : konstanta pegas
- x : perpindahan dari posisi kesetimbangan
<br> Osilasi harmonik secara sederhana dapat dihitung menggunakan gerak diferensial:
$$ m \frac{d^2 x}{dt^2} = -kx $$
Dimana jika menggunakan metode euler akan di pecah menjadi dua persamaan diferensial orde pertama:
1. $$ \frac{dx}{dt} = v $$
2. $$ \frac{dv}{dt} = \frac{-kx}{m}$$


Algoritma
1. Tebak nilai k, timestep (t), posisi awal (x0) dan kecepatan awal (v0)
2. Lakukan perhitungan dengan rumus:
$$ x_{n+1} = x +v \Delta t $$
$$ v_{n+1} = v + (\frac{-kx}{m}) \Delta t $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Initial value
k = 1.0
m = 1.0 
x0 = 1.0
v0 = 0.0 
time = np.linspace(100, 0, 1000)

# Initial array to store data
x = np.zeros(len(time))
v = np.zeros(len(time))
t = np.zeros(len(time))

def euler(x0, v0, time):
    x[0] = x0
    v[0] = v0
    for i in range(1, len(time)):
        x[i] = x[i-1] + v[i-1] * (time[i]-time[i-1])
        v[i] = v[i-1] + (-k/m * x[i-1]) * (time[i]-time[i-1])
        t[i] = t[i-1] + (time[i]-time[i-1])
    return x, t
x, t = euler(x0, v0, time)

# Plot
plt.plot(t, x)
plt.title('Osilasi Harmonik Menggunakan Metode Euler')
plt.xlabel('Waktu')
plt.ylabel('Posisi')
plt.show()
```
