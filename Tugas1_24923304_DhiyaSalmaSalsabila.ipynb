## **Jawab**

Algoritma mencari nilai akar dengan pendekatan gradien garis
1. Tebak nilai awal x dan nilai batas toleransi/error
2. Dengan rumus gradien
$$ y = mx + c$$
$$ y = y' (x-x_n) + y(x_n) $$
&emsp; untuk nilai akar maka nilai y = 0,
$$ 0 = y' (x-x_n) + y(x_n) $$
&emsp; maka
$$ x_n = x_{n-1} - \frac{y(x_n)}{y'(x_n)}

```python
# Initial value
y = lambda x: x*x - 5*x +6
y1 = lambda x: 2*x - 5
x = 1
err_ = 1e-10
```

```python
# Dengan fungsi for
n = 7 # Tentukan batas step
for i in range(n):
    x = abs(x - (y(x)/y1(x)))
    print("Nilai akar: ", x)
```

```python
# Dengan fungsi while
while abs(y(x))>err_:
    x = abs(x - (y(x)/y1(x)))
    print("Nilai akar: ", x)
```

```python
# Dengan fungsi rekursif
def integral(y, y1, x, e):
    x = abs(x - (y(x)/y1(x)))
    while abs(y(x))>e:
        print("Nilai akar: ", x)
        return integral(y, y1, x, err_)
hasil = integral(y, y1, x, err_)
print(hasil)
```

Algoritma hasil integral bentuk geometri trapesium
1. Tentukan nilai batas awal (a), batas akhir (b), dan selang (n)
2. Buat sebuah array/list untuk mengisi interval dari batas awal (a) ke batas akhir (b) dengan selang (n), hal ini untuk membagi luas di bawah kurva menjadi beberapa trapesium
3. Hitung panjang tiap trapesium tadi dengan rumus 
$$ h = \frac{(b-a)}{(n-1)} $$
4. Kemudian hitung luas trapesium dengan rumus
$$ luas = \frac{alas atas + alas bawah}{2} tinggi $$
&emsp; dengan batas awal (a), batas akhir (b), maka seluruh trapesium dihitung luas dengan
$$ luas = \frac{f(a)+f(x1)}{2} h + \frac{f(x1)+f(x2)}{2} h + \frac{f(x2)+f(x3)}{2} h + \frac{f(x3)+f(xb)}{2} h$$
$$ luas = \frac{h}{2} (f(a)+f(x1)+f(x1)+f(x2)+f(x2)+f(x3)+f(x3)+f(b))$$
$$ luas = \frac{h}{2} f(a)+2 \sum_{i=i+1}^{n-1} f(xi) +f(b))$$

```python
import numpy as np

# Batas integral dari 0 sampai pi/2 dengan selang 100
a = 0
b = np.pi/2
n = 100
x = np.linspace(a,b,n)

# Lebar tiap trapesium
h = (b-a)/(n-1)

# Fungsi sinus
f = np.sin(x)

# Integral bentuk geometri trapesium
hasilintegral = h/2*(f[0]+2*sum(f[1:n-1])+f[-1])
print("Luas daerah dibawah grafik adalah : ", hasilintegral)
```

### Plot grafik

```python
import matplotlib.pyplot as plt

# Plot sinus dari 0 sampai 90
plt.plot(x, f, label=r'$f(x) = \sin(x)$', color='r')

# Plot trapesium
for i in range(n - 1):
    # Buat trapesium dari dua titik (x[i], f[i]) dan (x[i+1], f[i+1])
    x_trap = [x[i], x[i+1], x[i+1], x[i]]
    y_trap = [0, 0, f[i+1], f[i]]
    plt.fill(x_trap, y_trap, 'g', edgecolor='g', alpha=0.5)

# Legend
plt.title(f"Grafik Sinus dari 0 sampai 90\n Luas = {hasilintegral}")
plt.xlabel('x (radian)')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.figure().set_figwidth(15)
plt.show()
```
